# Data cleaning

```python
# our age and cabin class having NA values so instead of dropping agedata rows we are going to fill na values with mean avg age.

# for that we we can check with passenger class relation

plt.figure(figsize=(12,7))
sns.boxplot(x="Pclass",y="Age",data=df1,palette=None)
```

```python
# creating function so we can change Age columns na values on basis of Pclass values
def cal_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    
    if pd.isnull(Age):
        
        if Pclass ==1:
            return 37
        
        elif Pclass ==2:
            return 29
        
        else:
            return 24
        
    else:
        return Age
```

```python
#apply function
df1['Age']=df1[['Age','Pclass']].apply(cal_age,axis=1)
```

```python
#here our age colums values got change
sns.heatmap(df1.isnull(),yticklabels=False,cbar= False)
```

```python
# after drop  we can see cabin also got deleted from above graph after run again that plot
df1.drop('Cabin',axis=1,inplace=True) 
```

```python
# after drop cabin column 
sns.heatmap(df1.isnull(),yticklabels=False,cbar= False)
```

# One hot encoding

```python
# for machine learning we are not going to take categorical values directly so we are going to chane it into numeric value
# here sex,Embarked are caategorical value so we are going to convert into mumeric

sex = pd.get_dummies(df1['Sex'],drop_first=True)
embark= pd.get_dummies(df1['Embarked'],drop_first=True)
```

```python
sex.head()
```

```python
embark.head()
```

```python
# drop unnecessory columns
df2=df1.drop(['Name','Sex','Ticket','Embarked'],axis='columns')
df2.head(20)
```

```python
# concat above converted categorical columns data eg. sex and embark columns

df3 = pd.concat([df2,sex,embark],axis='columns')
df3.head()
```

# Now we can start to build model here survived is our dependent variable and other all are independent variable so wiil drop surived column

```python
# for train data set we will drop Survived
X = df3.drop('Survived',axis='columns')
X.head()
```

```python
y= df3.Survived
y.head()
```

```python
# import train test model from sklearn
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.3)
```

```python
X_train
```

# Apply logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

# Train the model

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
# accuracy ouput 
score=model.score(X_test,y_test)
score
```
